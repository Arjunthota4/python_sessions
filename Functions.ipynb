```python
def biggest(x1,x2,x3):
    if x1>x2 and x1>x3:
        print("x1 is greatest")
        return(x1)
    elif x2>x3:
        print("x2 is greatest")
        return(x2)
    else:
        print("x3 is greatest")
        return(x3)
```

```python
y1=biggest(30,15,20)
print(y1)
```

```python
def add(a,b):
    return(a+b)
def mul(a,b):
    return(a*b)
def sub(a,b):
    return(a-b)
def div(a,b):
    return(a/b)
val_add=add(20,30)
val_mul=mul(20,30)
val_sub=sub(20,30)
val_div=div(50,10)
print(val_add)
print(val_mul)
print(val_sub)
print(val_div)
```
